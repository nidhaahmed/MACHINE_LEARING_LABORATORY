```python
!pip install numpy
import numpy as np
l1 = [1,2,3]
l2 = [[1,2,3],[4,5,6]]
a1 = np.array(l1)
a2 = np.array(l2)
s1 = a1.shape
s2 = a2.shape
print(s1)
print(s2)
```

```python
#Addition & Subtraction
aa = a1 + a2
sa = a1 - a2
print(aa)
print(sa)
print(aa.shape)
print(sa.shape)
```

```python
#SQRT & MIN & SUM & MAX & MEAN
print(np.sqrt(a1),'\n', np.sqrt(a2))
print(a1.sum(),'\n', a2.sum())
print(a1.min(),'\n', a2.min())
print(a1.max(),'\n', a2.max())
print(np.mean(a1),'\n', np.mean(a2))
```

```python
#Individual axis sum
a3 = np.array([[1,2,3],[4,5,6]])
print(np.sum(a3, axis=0))
print(np.sum(a3, axis=1))
```

```python
#MATRIX MULTIPLICATION
'''we use a function called matmul from numpy module'''
l4 = [1,2,3]
l5 = [[1,2,3],[4,5,6],[7,8,10]]
m = np.matmul(l4,l5)
print(m)
print("Shape : ", m.shape)
```

```python
#MATRIX INVERSE
'''we use a function called inv from numpy module
or we can also use np.linalg.inv'''
print(np.linalg.inv(l5))
```
